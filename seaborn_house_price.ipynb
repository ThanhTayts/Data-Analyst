```python
# import thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# đọc dữ liệu từ file csv
df = pd.read_excel('house_price_dống-da.xlsx')
df.head()
```

#### Vẽ biểu đồ xu hướng phân tích mối quan hệ giữa giá nhà với diện tích, giá nhà với số lượng phòng ngủ, nhận xét

```python
df.columns
```

```python
# biểu đồ xu hướng giữa giá nhà và diện tích
sns.lmplot(x="area", y="price",data = df);
```

```python
# Giá nhà cao không hoàn toàn phụ thuộc vào diện tích mà còn phụ thuộc vào nhiều yếu tố khác
```

```python
# biểu đồ xu hướng giữa giá nhà và số lượng phòng ngủ
sns.lmplot(x="bedroom", y="price",data = df);
```

```python
# tượng tự đối với số phòng ngủ cũng như vậy
```

#### Vẽ biểu đồ phân bố thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
# Biểu đồ phân bố của giá nhà theo các hướng
df.dropna()
plt.figure(figsize=(10,10))
sns.violinplot(x='house_direction',y = "price", data=df);
```

```python
# phân bố giá nhà theo các hướng không đồng đều
```

#### Vẽ biểu đồ tần số để đếm số nhà ở mỗi hướng nhà, nhận xét.

```python
#vẽ biểu đồ tần số đếm số nhà ở mỗi hướng
plt.figure(figsize=(10,10))
sns.countplot(x = "house_direction", data = df);
```

```python
# Ở hướng Đông Nam có số lượng nhà nhiều nhất và ở hướng Bắc có số lượng nhà ít nhất
```

#### Vẽ biểu đồ boxplot thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
# Vẽ biểu đồ box plot thể hiện phân bố giá nhà theo hướng
plt.figure(figsize=(10,10))
sns.boxplot(x = "house_direction", y = "price", data=df);
```
