#### Yêu cầu 3

```python
# Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.
df2 = df[(df['land_certificate']=='Sổ đỏ') & (df['bedroom'] >= 3)]
df2[['address', 'price', 'house_direction', 'balcony_direction']].reset_index(drop=True)
```

#### Yêu cầu 4

```python
# Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.
df.groupby('type_of_land').agg({'price': ['mean', 'max', 'min']})
# df.groupby('type_of_land')['price'].agg(['mean', 'max', 'min'])
```

```python
# Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.
df.groupby('ward_name')['bedroom', 'toilet', 'floor'].agg(['mean', 'max', 'min'])
```
