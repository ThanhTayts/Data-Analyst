#### Yêu cầu 2

```python
# Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia
df_pivot = df.pivot_table(values='Quantity', index='InvoiceNo', columns='Country', aggfunc= 'mean')
df_pivot
```

#### Yêu cầu 3

```python
# Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
df_pivot = df.pivot_table(values='Quantity', index='CustomerID', columns='StockCode', aggfunc= {'max','min'})
df_pivot
```

#### Yều cầu 4

```python
# Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
df_pivot = df.pivot_table(values=['Quantity','UnitPrice'], index=['StockCode'], columns='Description', aggfunc= {'Quantity':np.sum,'UnitPrice':np.mean})
df_pivot
```

#### Yêu cầu 5 

```python
df['Date'] = df['InvoiceDate'].apply(lambda x: x.split(' ')[0])
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
# Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.
df_pivot = df.pivot_table(values='Quantity', index='Description', columns='Date', aggfunc= 'sum')
df_pivot
```
