```python
# import các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# đọc dữ liệu
df = pd.read_csv('D:\Desktop\Data_Analyst\Jupiter_Notebook_Pandas\Pandas\shopeep_koreantop_clothing_shop_data.csv')
df.head()
```

#### Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm

```python
df_join =  df.groupby('join_year')['name'].count()
df_join
```

```python
# Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm
plt.figure(figsize=(10,5))
plt.xlabel('Năm gia nhập')
plt.ylabel('Số lượng shop')
plt.title('Số lượng shop gia nhập theo các năm')
plt.bar(height= df_join,x = df_join.index, color='red', linestyle='dashed', linewidth=2);
```

#### Vẽ biểu đồ thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.

```python
plt.figure(figsize=(10,5))
plt.xlabel('Tỉ lệ phản hồi')
plt.ylabel('Số lượng khách hàng đánh giá tốt')
plt.title('Tỉ lệ phản hồi theo số lượng khách hàng đánh giá tốt ')
plt.scatter(df['response_rate'], df['rating_good'], color='blue');
```

#### Vẽ biểu đồ thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu

```python
plt.figure(figsize=(15,8))
plt.xlabel('Thời gian phản hồi')
plt.ylabel('Số lượng khách hàng đánh giá xấu')
plt.title('Thời gian phản hồi theo số lượng khách hàng đánh giá xấu ')
plt.scatter(df['response_time'],df['rating_bad'] , color='green');
```

#### Vẽ biểu đồ thể hiện xu hướng của số lượng shop gia nhập theo thời gian

```python
# biểu đồ xu hướng của số lượng shop gia nhập theo thời gian
df_join_day =  df.groupby('join_day')['name'].count()
df_join_month =  df.groupby('join_month')['name'].count()
df_join_year =  df.groupby('join_year')['name'].count()
```

```python
plt.figure(figsize=(15,5))
plt.xlabel('Thời gian gia nhập theo ngày')
plt.ylabel('Số lượng shop gia nhập')
plt.title('Biểu đồ xu hướng của số lượng shop gia nhập theo ngày')
plt.plot(df_join_day.index,df_join_day ,color='red', linestyle='dashed', linewidth=2,marker='o',markersize=10);
```

```python
plt.figure(figsize=(15,8))
plt.xlabel('Thời gian gia nhập theo tháng')
plt.ylabel('Số lượng shop gia nhập')
plt.title('Biểu đồ xu hướng của số lượng shop gia nhập theo tháng')
plt.plot(df_join_month.index,df_join_month ,color='green', linestyle='solid', linewidth=2,marker='o',markersize=10);
```

```python
plt.figure(figsize=(15,8))
plt.xlabel('Thời gian gia nhập theo năm')
plt.ylabel('Số lượng shop gia nhập')
plt.title('Biểu đồ xu hướng của số lượng shop gia nhập theo năm')
plt.plot(df_join_year.index,df_join_year ,color='blue', linestyle='dashdot', linewidth=2,marker='o',markersize=10);
```
